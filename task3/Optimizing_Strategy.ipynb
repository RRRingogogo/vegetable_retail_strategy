```python
import pandas as pd

# 从Excel文件中加载数据
file_path = '624-630单品.xlsx'
df = pd.read_excel(file_path)

# 假设预测的进货量存储在一个字典中，以分类名称为键，进货量为值
predicted_inventory = {
    '水生根茎类': 29.22246053,  # 以示例数据为准
    '花叶类': 186.2949681,
    '花菜类': 28.50318887,
    '茄类': 31.82545023,
    '辣椒类': 119.6600446,
    '食用菌': 72.10237528
}

# 创建一个字典来存储每个单品的填充情况，包括单品名称、填充销售量和销售利润
item_filling = {}

# 设置每个分类的最大商品数量
max_items_per_category = 7  # 假设每个分类最多10个商品

# 根据分类名称分组数据
grouped_data = df.groupby('分类名称')

# 存储每个分类的填充销售量
category_sales = {}

# 针对每个分类应用贪心算法
for category_name, category_data in grouped_data:
    # 根据单位利润从高到低排序
    category_data = category_data.sort_values(by='单位利润', ascending=False)
    
    # 初始化背包和其他变量
    backpack = []
    remaining_inventory = predicted_inventory[category_name]
    total_profit = 0
    
    # 使用贪心算法填充背包，同时考虑最大商品数量
    items_added = 0
    
    for index, row in category_data.iterrows():
        demand = row['日均销量(千克)']
        item_name = row['单品名称']
        profit = row['日均利润']
        
        if remaining_inventory >= demand and items_added < max_items_per_category:
            backpack.append((item_name, demand, profit))
            total_profit += profit
            remaining_inventory -= demand
            items_added += 1
    
    # 计算放大比例
    total_filled_sales = sum(item[1] for item in backpack)
    scale_factor = predicted_inventory[category_name] / total_filled_sales
    
    # 根据放大比例修改销售量和销售利润
    for i in range(len(backpack)):
        backpack[i] = (backpack[i][0], backpack[i][1] * scale_factor, backpack[i][2] * scale_factor)
    
    # 存储填充情况和总销售利润
    item_filling[category_name] = {'填充情况': backpack, '总销售利润': total_profit}
    category_sales[category_name] = sum(item[1] for item in backpack)

# 打印每个分类的单品填充情况和总销售利润
for category_name, result in item_filling.items():
    print(f"分类名称: {category_name}")
    print(f"单品填充情况:")
    for item in result['填充情况']:
        print(f"单品名称: {item[0]}, 填充销售量: {item[1]:.2f} 千克, 销售利润: {item[2]:.2f}")
    print(f"总销售利润: {result['总销售利润']:.2f}")
    print(f"填充销售总量: {category_sales[category_name]:.2f}")
    print("\n")
```

```python
import random
import numpy as np
import pandas as pd
import math

# 从Excel文件中加载数据
file_path = '单品及成本.xlsx'
df = pd.read_excel(file_path)

# 假设预测的进货量存储在一个字典中，以分类名称为键，进货量为值
predicted_inventory = {
    '水生根茎类': 29.22246053,  # 以示例数据为准
    '花叶类': 186.2949681,
    '花菜类': 28.50318887,
    '茄类': 31.82545023,
    '辣椒类': 119.6600446,
    '食用菌': 72.10237528
}

# 定义适应度函数：总销售利润
def fitness(individual):
    total_profit = 0
    selected_items = []
    
    for i, is_selected in enumerate(individual):
        if is_selected:
            row = df.iloc[i]
            demand = row['日均补货']
            category = row['分类名称']
            profit = row['利润率']
            
            if demand <= predicted_inventory[category]:
                total_profit += profit
                selected_items.append(i)
    
    # 添加一个约束条件，确保所选商品数量不超过33个
    if len(selected_items) <= 33:
        return total_profit
    else:
        return 0

# 初始化当前解和当前适应度
current_solution = [random.choice([0, 1]) for _ in range(len(df))]
current_fitness = fitness(current_solution)

# 设置模拟退火算法参数
initial_temperature = 1000
min_temperature = 0.1
cooling_rate = 0.995

# 模拟退火主循环
current_temperature = initial_temperature
while current_temperature > min_temperature:
    # 生成邻居解
    neighbor_solution = current_solution.copy()
    index_to_flip = random.randint(0, len(df) - 1)
    neighbor_solution[index_to_flip] = 1 - neighbor_solution[index_to_flip]
    neighbor_fitness = fitness(neighbor_solution)
    
    # 计算能量差
    energy_difference = neighbor_fitness - current_fitness
    
    # 如果邻居解更好或按一定概率接受差解，则更新当前解
    if energy_difference > 0 or random.random() < math.exp(energy_difference / current_temperature):
        current_solution = neighbor_solution
        current_fitness = neighbor_fitness
    
    # 降低温度
    current_temperature *= cooling_rate

# 打印结果
result_df = df.copy()
result_df['是否选中'] = current_solution
result_df = result_df[result_df['是否选中'] == 1]

print(result_df[['分类名称', '单品名称', '日均补货', '利润率']])
print("总销售利润:", current_fitness)
```

```python
import pandas as pd

# 从Excel文件中加载数据
file_path = '单品进货量.xlsx'
df = pd.read_excel(file_path)

# 假设预测的进货量存储在一个字典中，以分类名称为键，进货量为值
predicted_inventory = {
    '水生根茎类': 29.22246053,  # 以示例数据为准
    '花叶类': 186.2949681,
    '花菜类': 28.50318887,
    '茄类': 31.82545023,
    '辣椒类': 119.6600446,
    '食用菌': 72.10237528
}

# 计算每一类单品的日均补货总和和预测进货量总和
daily_replenishment_sum = df.groupby('分类名称')['日均补货'].sum()
predicted_inventory_sum = sum(predicted_inventory.values())

# 初始化一个字典，用于存储每一类单品的缩放因子
scaling_factors = {}

# 计算每一类单品的缩放因子
for category, replenishment_sum in daily_replenishment_sum.items():
    scaling_factor = predicted_inventory[category] / replenishment_sum
    scaling_factors[category] = scaling_factor

# 对每一类单品的日均补货进行缩放并替换原数据
for category, scaling_factor in scaling_factors.items():
    category_mask = df['分类名称'] == category
    df.loc[category_mask, '日均补货'] *= scaling_factor

# 打印结果
print(df[['分类名称', '单品名称', '日均补货']])
```
